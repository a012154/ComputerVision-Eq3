## Importing Libraries <a class="anchor" id="libraries" ></a>

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from google.colab.patches import cv2_imshow # for image display
```

```python
url = "https://placekitten.com/500/300"
img = io.imread(url)  
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img2 = io.imread("https://placekitten.com/500/300",0)
cv2_imshow(img2)
```

```python
#function to see image
def imgshow(img=img,img1=img1,img2=img2, rows=1,columns=3):
    print(img.shape,img1.shape,img2.shape)
    fig = plt.figure(figsize=(15,8))
    #adds a subplot at 1st position
    fig.add_subplot(rows,columns,1)
    plt.title('BGR Format')
    plt.imshow(img)

    #adds a subplot at 2nd position
    fig.add_subplot(rows,columns,2)
    plt.title('RGB Format')
    plt.imshow(img1)

    #adds a subplot at 3rd position
    fig.add_subplot(rows,columns, 3)
    plt.title('BGR to GRAY')
    plt.imshow(img2,cmap="gray")
    plt.show()
    
imgshow(img=img,img1=img1,img2=img2, rows=1,columns=3)
```

## Erosion <a class="anchor" id="erosion" ></a>

Erosin primarily involves eroding(thinning) outer surface(foreground) of image
- Tt is suggested to have foreground as white

```python
BGR2GRAY = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(BGR2GRAY,cmap="gray")
plt.show()
```

Eroding boundy of image upto 3 level

```python
def plt_imshow(title, image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(title)
    plt.grid(False)
    plt.show()
    
for i in range(0,3):
    eroded = cv2.erode(BGR2GRAY.copy(), None, iterations=i+1)
    plt_imshow("Eroded {} times".format(i+1), eroded)
```

### Binarize Image

```python
binr = cv2.threshold(BGR2GRAY, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
image = cv2.cvtColor(binr, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

### We can Invert Image to keep foreground in white

```python
invert = cv2.bitwise_not(binr)
image = cv2.cvtColor(invert, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

Then, we can define a kernel of $5 \times 5$ to use in erosion

```python
kernel = np.ones((5,5), np.uint8) #check for 3*3 or 7*7 kernel size too
print(kernel)
```

```python
for i in range(0,3):
    eroded = cv2.erode(invert.copy(),kernel=kernel, iterations=i+1)
    plt_imshow("Eroded {} times".format(i+1), eroded)
```

## Dilation <a class="anchor" id="dilation" ></a>

Involves dilating outer surface (foreground) of image
- For joining broken parts of an image together

```python
#binarize image
binr = cv2.threshold(BGR2GRAY,0,255,cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
kernel = np.ones((3,3),np.uint8)
invert = cv2.bitwise_not(binr)

#dilate image
dilation = cv2.dilate(binr, kernel, iterations=1)
 
plt.imshow(dilation, cmap='gray')
plt.show()
```

We can apply series of dilations on inverted image using no kernel


```python
for i in range(0,3):
    dilated = cv2.dilate(invert.copy(),None,iterations= i+1)
    plt_imshow("Dilated {} times".format(i+1),dilated)
```

```python
for i in range(0,4):
    dilated = cv2.dilate(BGR2GRAY.copy(),None,iterations= i+1)
    plt_imshow("Dilated {} times".format(i+1),dilated)
```

## Opening <a class="anchor" id="opening" ></a>

Erosion followed by dilation in outer surface (foreground) of image
- Generally used to remove noise (small blobs) in image


```python
#opening image
opening = cv2.morphologyEx(binr,cv2.MORPH_OPEN,kernel,iterations=1)

#print output
plt.imshow(opening, cmap='gray')
plt.show()
```

```python
for i in range(0,3):
    opening = cv2.morphologyEx(binr.copy(),cv2.MORPH_OPEN,None,iterations= i+1)
    plt_imshow("Dilated {} times".format(i+1),opening)
    plt.show()
```

## Closing <a class="anchor" id="closing" ></a>

Dilation followed by erosion in outer surface (foreground) of image
- Generally used to close holes inside of objects or for connecting components together

```python
#closing image
closing = cv2.morphologyEx(binr, cv2.MORPH_CLOSE, kernel, iterations=1)
 
#print output
plt.imshow(closing,cmap='gray')
plt.show()
```

```python
for i in range(0,3):
    closing = cv2.morphologyEx(binr.copy(),cv2.MORPH_CLOSE,None,iterations= i+1)
    plt_imshow("Dilated {} times".format(i+1),closing)
    plt.show()
```

Iterating with different kernels

```python
kernelSizes = [(3,3), (5,5), (7,7)]
BGR2GRAY = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

for kernelSize in kernelSizes:
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, kernelSize)
    closing = cv2.morphologyEx(BGR2GRAY, cv2.MORPH_CLOSE, kernel)
    plt_imshow("Closing Kernel Size: ({}, {})".format(kernelSize[0], kernelSize[1]), closing)
```

## Morphological Gradient <a class="anchor" id="gradient" ></a>

It first applies erosion and dilation individually on image and then computes difference between eroded and dilated image

```python
# use morph gradient
morph_gradient = cv2.morphologyEx(invert,cv2.MORPH_GRADIENT,kernel)
 
#print output
plt.imshow(morph_gradient, cmap='gray')
plt.show()
```

```python
for i in range(0,3): 
    # use morph gradient
    morph_gradient = cv2.morphologyEx(invert,cv2.MORPH_GRADIENT,kernel+i)
    #print output
    plt.imshow(morph_gradient, cmap='gray')
    plt.show()
```

Iterating with different kernels

```python
kernelSizes = [(3,3), (5,5), (7,7)]
BGR2GRAY = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

for kernelSize in kernelSizes:
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, kernelSize)
    gradient = cv2.morphologyEx(BGR2GRAY, cv2.MORPH_GRADIENT, kernel)
    plt_imshow("Gradient: ({}, {})".format(
        kernelSize[0], kernelSize[1]), gradient)
```

## Morphological Hats <a class="anchor" id="hats" ></a>

1. Top or White Hat
2. Black hat

### Top or White Hat <a class="anchor" id="top" ></a>

It is the difference between original (grayscale/single channel) input image and opening image

- Finds light(bright) regions of an image on dark background


```python
#construct a kernel (13x5)
kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(13,5))
kernel
```

```python
tophat = cv2.morphologyEx(BGR2GRAY, cv2.MORPH_TOPHAT, kernel)

plt.imshow(tophat, cmap='gray')
plt.show()
```

### Black Hat <a class="anchor" id="black" ></a>

Output is a difference between input image & opened image
- Finds dark regions on a light background


```python
blackhat = cv2.morphologyEx(BGR2GRAY,cv2.MORPH_BLACKHAT,kernel)

plt.imshow(blackhat, cmap='gray')
plt.show()
```

#**Ejercicio 1: Erosión**

```python
#Get image:
url = "https://i.postimg.cc/Zn4dzS0X/img-botones-1.png"
img0 = io.imread(url)  

#Change ir to grayscale:
img01 = cv2.cvtColor(img0, cv2.COLOR_BGR2GRAY)

#Convert to binary image:
binr1 = cv2.threshold(img01, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
img02 = cv2.cvtColor(binr1, cv2.COLOR_BGR2RGB)
img = cv2.bitwise_not(img02)

#Apply erosion:
def erosion(img, kernel_intensity=5):
  # new img
  erosion_img = np.zeros_like(img)
  # kernel
  kernel = np.ones((kernel_intensity,kernel_intensity), np.uint8)

  for i in range(img.shape[0] - kernel.shape[0] + 1):
    for j in range(img.shape[1] - kernel.shape[1] + 1):
      temp_array = img[i:i+kernel.shape[0], j:j+kernel.shape[1]]
      erosion_img[i + int(kernel.shape[0]/2), j + int(kernel.shape[1]/2)] = np.min(temp_array)
  return erosion_img

#Create eroded images with different kernel sizes:
erosion_img_1 = erosion(img,5)
erosion_img_2 = erosion(img,10)
erosion_img_3 = erosion(img,25)

#Display images:
fig = plt.figure(figsize=(12, 8))
images = [img, erosion_img_1, erosion_img_2, erosion_img_3]
titles = ["Original Image", "Erosion - kernel (5,5)", "Erosion - kernel (10,10)", "Erosion - kernel (25,25)"]
for i in range(4):
  plt.subplot(1,4,i+1), plt.imshow(images[i], 'gray')
  plt.title(titles[i])
  plt.xticks([]),plt.yticks([])
plt.show()
```

Ahora veamos la diferencia entre los resultados cambiando el número de iterationes en las que se aplica la Erosión utilizando OpenCV.

```python
#OpenCV erosion:
eroded_1 = cv2.erode(img.copy(), None, iterations=1)
eroded_2 = cv2.erode(img.copy(), None, iterations=2)
eroded_3 = cv2.erode(img.copy(), None, iterations=3)
eroded_4 = cv2.erode(img.copy(), None, iterations=4)

fig = plt.figure(figsize=(15, 9))
images = [img, eroded_1, eroded_2, eroded_3, eroded_4]
titles = ["Original Image", "CV Erosion - 1 iteration", "CV Erosion - 2 iterations", "CV Erosion - 3 iterations", "CV Erosion - 4 iterations"]
for i in range(5):
  plt.subplot(1,5,i+1), plt.imshow(images[i], 'gray')
  plt.title(titles[i])
  plt.xticks([]),plt.yticks([])
plt.show()
```

##Comentarios

La erosión nos permite enfocar las características de un objeto específico dentro de una imagen, simplificando la estructura y descomponiéndolo en partes.
Por lo que resulta muy útil permitir separar objetos que están muy juntos en alguna línea de inspección de componentes. 

#**Ejercicio 2: Dilatación**

```python
#url = "https://i.postimg.cc/pd4hQ0fF/ezgif-com-gif-maker-1.jpg"
```

```python
#Get image:
url = "https://i.postimg.cc/kXZxYDWp/img-carta-a-mano-1.png"
img0 = io.imread(url)  
img01 = cv2.cvtColor(img0, cv2.COLOR_BGR2GRAY)

#Convert to binary image:
binr1 = cv2.threshold(img01, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
img02 = cv2.cvtColor(binr1, cv2.COLOR_BGR2RGB)
img = cv2.bitwise_not(img02)

#Apply dilation
def dilation(img, kernel_intensity=5):
  dilate_img = np.zeros_like(img)
  kernel = np.ones((kernel_intensity,kernel_intensity), np.uint8)
  for i in range(img.shape[0] - kernel.shape[0] + 1):
    for j in range(img.shape[1] - kernel.shape[1] + 1):
      temp_array = img[i:i+kernel.shape[0], j:j+kernel.shape[1]]
      dilate_img[i + int(kernel.shape[0]/2), j + int(kernel.shape[1]/2)] = np.max(temp_array)
  return dilate_img

dilate_img_1 = dilation(img,2)
dilate_img_2 = dilation(img,3)
dilate_img_3 = dilation(img,4)

#Display images:
fig = plt.figure(figsize=(22, 15))
images = [img, dilate_img_1, dilate_img_2, dilate_img_3]
titles = ["Original Image", "Dilation - kernel (2,2)", "Dilation - kernel (3,3)", "Dilation - kernel (4,4)"]
for i in range(4):
  plt.subplot(1,4,i+1), plt.imshow(images[i], 'gray')
  plt.title(titles[i])
  plt.xticks([]),plt.yticks([])
plt.show()
```

```python
#OpenCV dilation:
dilated_1 = cv2.dilate(img.copy(), None, iterations=1)
dilated_2 = cv2.dilate(img.copy(), None, iterations=2)
dilated_3 = cv2.dilate(img.copy(), None, iterations=3)
dilated_4 = cv2.dilate(img.copy(), None, iterations=4)

fig = plt.figure(figsize=(24, 18))
images = [img, dilated_1, dilated_2, dilated_3, dilated_4]
titles = ["Original Image", "CV Dilation - 1 iteration", "CV Dilation - 2 iterations", "CV Dilation - 3 iterations", "CV Dilation - 4 iterations"]
for i in range(5):
  plt.subplot(1,5,i+1), plt.imshow(images[i], 'gray')
  plt.title(titles[i])
  plt.xticks([]),plt.yticks([])
plt.show()
```

##Comentarios

Con la dilatación podemos expandir el contorno de un objeto, con este tipo de operación debemos establecer/conocer el límite y no deformar la imagen y que sea ilegible. De manera que resulta muy útil para poder leer una imagen de algún texto escrito a mano.

#**Ejercicio 3: Opening y Closing**

```python
url = "https://i.postimg.cc/pXJs0j1T/ezgif-com-gif-maker-2.jpg"
img = io.imread(url)  
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


def morph_open(img, kernel_intensity=5):
  temp_img = erosion(img, kernel_intensity)
  temp_img = dilation(img, kernel_intensity)
  return temp_img

def morph_close(img, kernel_intensity=5):
  temp_img = dilation(img, kernel_intensity)
  temp_img = erosion(img, kernel_intensity)
  return temp_img

morph_open_img = morph_open(img,3)
morph_close_img = morph_close(img,2)

# display imgs
final_frame = cv2.hconcat((img, morph_open_img, morph_close_img))
cv2_imshow(final_frame)
```

##Comentarios

A través de estas operaciones podemos obtener dos distintos resultados:

1.- Open:  Nos sirve para poder enforcarnos en un objeto específico (elimina los pixeles que no queremos), aplica primero la operación de erosión y posterior dilatación.

2.- Closing: Nos sirve para poder enforcarnos en un objeto específico (agrega los pixeles que queremos), aplica primero la operación de dilatación y posterior erosión.
